# A simple linear regression fit of a Neural Network using Tensorflow and Keras.

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## Building a connected layer of a Neural Network, a sequential model.

```python
model= tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
```

## Compiling the model, to add a loss function and an optimiser.

```python
model.compile(loss='mean_squared_error', optimizer= 'sgd');
```

## Dataset

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

## Fitting the Neural Network

```python
model.fit(xs, ys, epochs=1000)
```

```python
print(model.predict([10]))
```
